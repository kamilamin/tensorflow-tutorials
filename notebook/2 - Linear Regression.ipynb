```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

rent = pd.read_csv('rentalprice.csv')
```

```python
rent.head()
```

```python
print(rent.shape)
```

```python
#load data
room = rent["Bedroom"] 
cost = rent["Cost"]
rent.describe()
```

```python
plt.plot(room,cost,'ro')
plt.axis([0,10000,0,60000])
plt.show()
```

```python
import tensorflow as tf

slope = tf.Variable([0.3])
intercept = tf.Variable([0.3])
bedroomsize = tf.placeholder(tf.float32)

actual_value =  tf.placeholder(tf.float32)

#Linear model
predicted_value = slope * bedroomsize + intercept
```

```python
#cost
costforstep = tf.reduce_sum(tf.square(predicted_value - actual_value))
```

```python
#optimizer
optimizer = tf.train.GradientDescentOptimizer(0.001)
train = optimizer.minimize(costforstep)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

#loop
for i in range(1000):
    data = {bedroomsize: room, actual_value:cost}
    sess.run(train, feed_dict=data)
    c_slope, c_intercept, c_cost = sess.run([slope,intercept,costforstep], feed_dict=data)
    print("Slope: %s Intercept: %s Cost: %s"%(c_slope,c_intercept,c_cost))
```
